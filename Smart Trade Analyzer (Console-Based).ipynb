````python

class Trade:

  #constructor
  def __init__(self, symbol, buy_price, sell_price, quantity):

    if buy_price <= 0 or sell_price <= 0:
      raise ValueError("The prices must be positive.")

    if quantity <= 0:
      raise ValueError("The quantity must be positive.")

    self.symbol = symbol
    self.buy_price = buy_price
    self.sell_price = sell_price
    self.quantity = quantity

  #calculte profite
  def profit(self):
    return (self.sell_price - self.buy_price) * self.quantity


  #checking status
  def status(self):
    if self.profit() > 0:
      return "PROFIT"

    elif self.profit() < 0:
      return "LOSS"
    else:
      return "BREAKEVEN"

  def __str__(self):
    return f"{self.symbol} | QTY: {self.quantity}" f"P/L: {self.profit()} | Status: {self.status()}"

````

````python
class TradeAnalyzer:
  #constructor
  def __init__(self):
    self.trades = []

  def add_trade(self, trade):
    self.trades.append(trade)

  def total_profit(self):
    total = 0
    for trade in self.trades:
      total += trade.profit()
    return total


  def best_trade(self):
    # max_p = float("-inf")
    # best_trade = None
    # for trade in self.trades():
    #   if trade.profit() > max_p:
    #     max_p = trade.profit()
    #     best_trade = trade
    # return best_trade

    return max(self.trades, key=lambda t: t.profit())


  def worst_trade(self):

    # min_p = float("inf")
    # worst_trade = None
    # for trade in self.trades:
    #   if trade.profit() < min_p:
    #     min_p = trade.profit()
    #     worst_trade = trade

    # return worst_trade

    return min(self.trades, key=lambda t: t.profit())

  def summary(self):
    print("-" * 30)
    print("Trade Summary")
    print("-" * 30)
    for trade in self.trades:
      print(trade)
    print("-" * 30)
    print(f"Total Profit: {self.total_profit()}")
    print(f"Best Trade: {self.best_trade()}")
    print(f"Worst Trade: {self.worst_trade()}")
    print("-" * 30)


````

````python
class StrategyAnalyzer(TradeAnalyzer):

  def profitable_trades(self, min_profit):
    # p_trade = []
    # for trade in self.trades:
    #   if trade.profit() >= min_profit:
    #     p_trade.append(trade)
    # return p_trade

    # return [trade for trade in self.trades if trade.profit() >= min_profit]

    return list(filter(lambda t: t.profit() >= min_profit, self.trades))

  def loss_trades(self, max_loss):
    # l_trade = []
    # for trade in self.trades:
    #   if trade.profit() <= max_loss:
    #     l_trade.append(trade)
    # return l_trade

    # return [trade for trade in self.trades if trade.profit() <= max_loss]

    return list(filter(lambda t: t.profit() <= max_loss, self.trades))

````

````python
def main():
  analyzer = StrategyAnalyzer()

  #sample market data
  market_data = [
      ('AAPL', 150, 160, 10),
      ('GOOGL', 720, 690, 5),
      ('MSFT', 100, 100, 8),
      ('TSLA', 2800, 2950, 4),
      ('AMZN', 1800, 1750, 6),
      ('NTLX', 342, 400, 5)
  ]

  for data in market_data:
    # symbol, buy_price, sell_price, quantity = data
    # trade = Trade(symbol, buy_price, sell_price, quantity)
    trade = Trade(*data)
    analyzer.add_trade(trade)

  analyzer.summary()

  print('\n Profitable Trades (> 100): ')
  for trade in analyzer.profitable_trades(100):
    print(trade)

  print('\n Lossy Trades (< -100): ')
  for trade in analyzer.loss_trades(-100):
    print(trade)


# main()
if __name__ == "__main__":
  main()
````
